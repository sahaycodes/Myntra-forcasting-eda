```python
def plot_top_categories(data, column, top_n=10):
    # Get the value counts of the column
    value_counts = data[column].value_counts()
    
    # Identify top N categories
    top_categories = value_counts.nlargest(top_n).index
    
    # Create a new column for plotting
    data[f'{column}_top'] = data[column].apply(lambda x: x if x in top_categories else 'Other')

# Plot the count plot
    plt.figure(figsize=(12, 6))
    sns.countplot(y=data[f'{column}_top'], order=data[f'{column}_top'].value_counts().index)
    plt.title(f'Top {top_n} {column.capitalize()}s')
    plt.xlabel('Count')
    plt.ylabel(column.capitalize())
    plt.show()
    
```

```python
# Plot count plots for the categorical columns
categorical_columns = ['seller', 'name']  # Add other categorical columns as needed
for col in categorical_columns:
    plot_top_categories(data, col, top_n=10)  # Adjust top_n as needed
```

```python
import pandas as pd

# Assuming 'data' is your DataFrame with the columns specified
# Convert 'random_date' to datetime format
data['random_date'] = pd.to_datetime(data['random_date'])

# Creating time-based features
data['year'] = data['random_date'].dt.year
data['month'] = data['random_date'].dt.month
data['day'] = data['random_date'].dt.day
data['dayofweek'] = data['random_date'].dt.dayofweek

# Creating lag features for numerical columns
numerical_cols = ['price', 'mrp', 'rating', 'ratingTotal', 'discount']
for col in numerical_cols:
    data[f'{col}_lag_1'] = data[col].shift(1)
    data[f'{col}_lag_7'] = data[col].shift(7)

# Creating rolling average features for numerical columns
for col in numerical_cols:
    data[f'{col}_rolling_mean_7'] = data[col].rolling(window=7, min_periods=1).mean()
    data[f'{col}_rolling_mean_30'] = data[col].rolling(window=30, min_periods=1).mean()

# Drop rows with NaN values created by lag/rolling features
data.dropna(inplace=True)

# Display the updated DataFrame with new features
print(data.head())
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pickle

# Assuming you have already created the features and target variables as specified

# Replace 'demand' with 'rating' as the target variable
target_attribute = 'rating'

# Split the data into training and testing sets
X_train = train_data[['year', 'month', 'day', 'dayofweek', 
                      'price_lag_1', 'price_lag_7', 
                      'price_rolling_mean_7', 'price_rolling_mean_30',
                      'mrp_lag_1', 'mrp_lag_7',
                      'mrp_rolling_mean_7', 'mrp_rolling_mean_30',
                      'rating_lag_1', 'rating_lag_7',
                      'rating_rolling_mean_7', 'rating_rolling_mean_30',
                      'ratingTotal_lag_1', 'ratingTotal_lag_7',
                      'ratingTotal_rolling_mean_7', 'ratingTotal_rolling_mean_30',
                      'discount_lag_1', 'discount_lag_7',
                      'discount_rolling_mean_7', 'discount_rolling_mean_30']]
y_train = train_data[target_attribute]

X_test = test_data[['year', 'month', 'day', 'dayofweek', 
                    'price_lag_1', 'price_lag_7', 
                    'price_rolling_mean_7', 'price_rolling_mean_30',
                    'mrp_lag_1', 'mrp_lag_7',
                    'mrp_rolling_mean_7', 'mrp_rolling_mean_30',
                    'rating_lag_1', 'rating_lag_7',
                    'rating_rolling_mean_7', 'rating_rolling_mean_30',
                    'ratingTotal_lag_1', 'ratingTotal_lag_7',
                    'ratingTotal_rolling_mean_7', 'ratingTotal_rolling_mean_30',
                    'discount_lag_1', 'discount_lag_7',
                    'discount_rolling_mean_7', 'discount_rolling_mean_30']]
y_test = test_data[target_attribute]

# Initialize and train the XGBoost model
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Save the model to a .pkl file
model_filename = 'xgboost_model.pkl'
with open(model_filename, 'wb') as file:
    pickle.dump(model, file)

# Optionally, you can also plot actual vs. predicted values
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.3)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title(f'Actual vs. Predicted {target_attribute}')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

# Initialize the GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', verbose=1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Print the best parameters and score
print(f'Best parameters: {grid_search.best_params_}')
print(f'Best score: {grid_search.best_score_}')
```
